```python
import pandas as pd
import json
```

```python
with open('./data/file_4.json') as json_file:  
    data = json.load(json_file)
```

```python
titles = []
authors = []
dates = []
ratings = []
verifieds =[]
texts = []
```

```python
for i, page in enumerate(data):
    if i == 0:
        date_scraped = data['date_scraped']
    else:
        all_reviews = data[page]['reviews']
        for review in all_reviews:
            title = review['title']
            author = review['author']
            date = review['date']
            rating = review['rating']
            try:
                verified = review['verified_buy']
            except:
                verified = "NONE"
            text = review['body']

            titles.append(title)
            authors.append(author)
            dates.append(date)
            ratings.append(rating)
            verifieds.append(verified)
            texts.append(text)
```

```python
df = pd.DataFrame({
    "Date" : dates,
    'Rating': ratings,
    "Title" : titles,
    "Author" : authors,
    "Review": texts,
    "Verified": verifieds
})
```
